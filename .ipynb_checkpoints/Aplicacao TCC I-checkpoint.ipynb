## **Refinamento dos datasets**

*   Normalização
*   Gerar novos dataframes
*   Salvar dados em variaveis genericas
*   Separar os conjuntos dependente e independente



```python
def prepara_dados(dependente, independente):
    df = pd.read_csv('/content/drive/My Drive/TCC/_TABELA_FINAL.csv', sep=';', encoding='latin-1')
    #construir o dataframe
    # data = df[[dependente] + independente]
    data = df.groupby('id')[[dependente] + independente].mean()
    # data = df.groupby('id_microrregiao')[[dependente] + independente].mean()
    # data = data[data['totalBeneficiarios'] < 30000]

    # redimensionar os dados
    scaler = MinMaxScaler().fit(data)
    data_norm = pd.DataFrame(scaler.transform(data), columns=data.columns)

    # separar os dados em dependente e independentes
    X = data_norm.drop(columns=[dependente])
    y = data_norm[dependente]
    return X, y
```

## **MODELO MLP**

```python
def build_mlp(input_dim):
    model = keras.Sequential([
        layers.Dense(8, activation='relu', input_shape=(input_dim,)),
        layers.Dense(4, activation='relu'),
        layers.Dense(2, activation='relu'),
        layers.Dense(1)  # saída para regressão
    ])
    #Definindo o otimizador
    optimizer = tf.keras.optimizers.RMSprop(
        learning_rate = 0.001  #Taxa de aprendizado de 0,001
    )

    #Mean Squared Error (MSE) é a função de perda padrão em modelos de regressão
    model.compile(loss = 'mse',  #Função de perda: erro quadrático médio
                  optimizer = optimizer,  #Otimizador RMSprop
                  metrics = ['mse'])
    return model
```

```python
dpendentes_vars = ['totalBeneficiarios','valorTotalRepassado']
for dependente in dpendentes_vars:
  blocos = [
      ['Idese'],
      [ 'Bloco Educação\Ensino Fundamental\Anos Finais EF',
        'Bloco Educação\Ensino Fundamental\Anos Iniciais EF',
        'Bloco Educação\Ensino Fundamental', 'Bloco Educação\Ensino Médio',
        'Bloco Educação\Escolaridade Adulta', 'Bloco Educação\Pré Escola'],
      ['Bloco Renda\Apropriação da Renda',
        'Bloco Renda\Geração da Renda'],
      ['Bloco Saúde\Condições Gerais de Saúde\Mortes por Causas Evitáveis',
        'Bloco Saúde\Condições Gerais de Saúde',
        'Bloco Saúde\Condições Gerais de Saúde\Óbitos por Causas Mal Definidas',
        'Bloco Saúde\Longevidade',
        'Bloco Saúde\Saúde Materno Infantil\Consultas Pré Natal',
        'Bloco Saúde\Saúde Materno Infantil\Mortalidade de Menores de 5 anos',
        'Bloco Saúde\Saúde Materno Infantil']
      # [ 'Bloco Educação','Bloco Renda', 'Bloco Saúde'],
      # ['Bloco Educação'],
      # ['Bloco Renda'],
      # ['Bloco Saúde'],
      ] # variavel independente
  # blocos = [[ 'Bloco Educação','Bloco Renda', 'Bloco Saúde']] # variavel independente
  # independente = ['Bloco Renda', 'Bloco Saúde'] # variavel independente
  resultados_mlp = pd.DataFrame(columns=['Bloco','R2','Desvio R2','MSE','Desvio MSE'])
  historys = []

  for bloco in list(blocos):
    X, y = prepara_dados(dependente,bloco)

    r2_ = []
    r_ = []
    mse_ = []
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    input_dim = X.shape[1]

    for fold, (train_idx, test_idx) in enumerate(kf.split(X), 1):
      X_tr, X_te = X.iloc[train_idx], X.iloc[test_idx]
      y_tr, y_te = y.iloc[train_idx], y.iloc[test_idx]

      # Cria e treina um modelo novo para este fold
      model = build_mlp(input_dim)
      historys.append(model.fit(X_tr, y_tr, epochs=20, verbose=0))
      # Faz predição e calcula métricas
      y_pred = model.predict(X_te).flatten()
      r2_.append(r2_score(y_te, y_pred))
      mse_.append(mean_squared_error(y_te, y_pred))
      # print(f"Bloco {bloco} Fold {fold}")
      # print("Correlation Coefficient in training set: %.4f" % np.corrcoef(np.transpose(train_predictions), np.transpose(train_y))[0,1])

    #salvando os resultados para o atributo analisado
    resultados_mlp = pd.concat([resultados_mlp, pd.DataFrame([{
        'Bloco': bloco,
        'R2': np.mean(r2_),
        'Desvio R2' : np.std(r2_),
        'MSE': np.mean(mse_),
        'Desvio MSE' : np.mean(mse_)
        }])], ignore_index=True)

  print(f'\n{dependente}\n')
  print(resultados_mlp)
```

## **REGRESSÃO LINEAR**



```python
vars_dependentes = ['totalBeneficiarios', 'valorTotalRepassado']
for dependente in vars_dependentes:
  blocos = [
      ['Idese'],
      ['Bloco Educação'],
      ['Bloco Renda'],
      # ['Bloco Saúde\Condições Gerais de Saúde\Mortes por Causas Evitáveis',
      #   'Bloco Saúde\Condições Gerais de Saúde',
      #   'Bloco Saúde\Condições Gerais de Saúde\Óbitos por Causas Mal Definidas',
      #   'Bloco Saúde\Longevidade',
      #   'Bloco Saúde\Saúde Materno Infantil\Consultas Pré Natal',
      #   'Bloco Saúde\Saúde Materno Infantil\Mortalidade de Menores de 5 anos',
      #   'Bloco Saúde\Saúde Materno Infantil']
      ['Bloco Saúde']
  ]

  resultados_rl = pd.DataFrame(columns=['Bloco', 'R2','Desvio R2','MSE','Desvio MSE'])

  for bloco in blocos:
      X, y = prepara_dados(dependente,bloco)

      r2_scores = []
      r_ =[]
      mse_ = []
      kf = KFold(n_splits=5, shuffle=True, random_state=42)
      # kf = KFold(n_splits=5)

      for fold, (train_idx, test_idx) in enumerate(kf.split(X), 1):
          X_tr, X_te = X.iloc[train_idx], X.iloc[test_idx]
          y_tr, y_te = y.iloc[train_idx], y.iloc[test_idx]

          modelo = LinearRegression()
          modelo.fit(X_tr, y_tr)
          y_pred = modelo.predict(X_te)

          r2_scores.append(r2_score(y_te, y_pred))
          mse_.append(mean_squared_error(y_te, y_pred))
          # print(f"Bloco {bloco} Fold {fold} R2: {r2}")

      r2_medio = np.mean(r2_scores)
      mse_medio = np.mean(mse_)
      r2_desvio = np.std(r2_scores)
      mse_desvio = np.std(mse_)

      # Adiciona o melhor resultado ao DataFrame
      resultados_rl = pd.concat([
          resultados_rl,
          pd.DataFrame([{
              'Bloco': bloco,
              'R2': r2_medio,
              'Desvio R2' : r2_desvio,
              'MSE': mse_medio,
              'Desvio MSE' : mse_desvio
          }])
      ], ignore_index=True)

  # Exibe resultados
  print(f'\n{dependente}\n')
  print(resultados_rl.to_string(index=False))
```
